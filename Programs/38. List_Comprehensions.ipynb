# Solution:

```python
x = int(input())  
y = int(input())  
z = int(input())  
n = int(input())  

A = [[i, j, k] 
     for i in range(x+1) 
     for j in range(y+1) 
     for k in range(z+1) 
     if (i + j + k) != n]
print(A)
```

# Equivalent  to:

```python
A = []  # start with an empty list

for i in range(x+1):              # loop over i
    for j in range(y+1):      # loop over j
        for k in range(z+1):  # loop over k
            if (i + j + k) != n:  # filter k
                A.append([i, j, k])  # add valid triple
print(A)
```

# General Example:

```python
coords = [[i, j, k] 
          for i in range(3) if i % 2 == 0      # filter i
          for j in range(3) if j != 1          # filter j
          for k in range(3) if (i + j + k) != 2]  # filter k
print(coords)
```

# Equivalent to:

```python
coords = []
for i in range(3):
    if i % 2 == 0:          # filter i
        for j in range(3):
            if j != 1:      # filter j
                for k in range(3):
                    if (i + j + k) != 2:   # filter k
                        coords.append([i, j, k])

print(coords)
```
