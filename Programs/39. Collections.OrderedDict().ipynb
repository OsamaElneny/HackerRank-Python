# Solution:

```python
from collections import OrderedDict

def calculate_net_prices(n, items):
    net_prices = OrderedDict()

    for item_name, price in items:
        net_prices[item_name] = net_prices.get(item_name, 0) + int(price)

    return net_prices

n = int(input().strip())
items = [tuple(input().rsplit(" ", 1)) for _ in range(n)]
result = calculate_net_prices(n, items)
for name, total in result.items():
    print(f"{name} > {total}")
```
