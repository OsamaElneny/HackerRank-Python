# Solution:

```python
import numpy as np

def read_matrix(n, m):
    rows = [list(map(int, input().split())) for _ in range(n)]
    return np.array(rows, dtype=int)

def compute_column_sum_and_product(A):
    col_sum = np.sum(A, axis=0)
    col_prod = int(np.prod(col_sum))
    return col_sum, col_prod

def main():
    n, m = map(int, input().split())
    A = read_matrix(n, m)
    col_sum, col_prod = compute_column_sum_and_product(A)
    print(col_sum)
    print(col_prod)


main()
```
