```python
def set_symmetric_difference_count(english, french):

    return len(english ^ french)  # '^' is the symmetric difference operator

n_eng = int(input().strip())
english = set(map(int, input().split()))

n_fre = int(input().strip())
french = set(map(int, input().split()))

print(set_symmetric_difference_count(english, french))
```
