# Solution:

```python
import numpy as np

def compute_matrices(A, B):
   
    return (
        A + B,
        A - B,
        A * B,
        np.floor_divide(A, B),
        A % B,
        A ** B
    )

def main():
    n, m = map(int, input().split())
    A = np.array([list(map(int, input().split())) for _ in range(n)])
    B = np.array([list(map(int, input().split())) for _ in range(n)])

    results = compute_matrices(A, B)
    for mat in results:
        print(mat)

main()
```
