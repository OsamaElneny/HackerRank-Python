```python
def set_mutations(A, operations):
   
    for op, other in operations:
        # getattr allows us to call the set method dynamically
        getattr(A, op)(other)
    return sum(A)

n = int(input().strip())
A = set(map(int, input().split()))

m = int(input().strip())
operations = []
for _ in range(m):
    op, _ = input().split()
    other = set(map(int, input().split()))
    operations.append((op, other))

print(set_mutations(A, operations))
```
