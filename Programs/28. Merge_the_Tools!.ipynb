# Solution:

```python
def merge_the_tools(string, k):
   
    for i in range(0, len(string), k):
        substring = string[i:i+k]
        seen = set()
        result = []
        for ch in substring:
            if ch not in seen:
                seen.add(ch)
                result.append(ch)
        print("".join(result))

s = input().strip()
k = int(input().strip())
merge_the_tools(s, k)
```
