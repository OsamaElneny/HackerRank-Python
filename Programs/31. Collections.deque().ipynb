# Solution:

```python
from collections import deque

def process_deque_operations(n, operations):

    d = deque()

    for op in operations:
        parts = op.split()
        command = parts[0]

        if command == "append":
            d.append(parts[1])
        elif command == "appendleft":
            d.appendleft(parts[1])
        elif command == "pop":
            if d:  # avoid error if empty
                d.pop()
        elif command == "popleft":
            if d:  # avoid error if empty
                d.popleft()

    return list(d)

n = int(input().strip())
operations = [input().strip() for _ in range(n)]

result = process_deque_operations(n, operations)
print(*result)
```
