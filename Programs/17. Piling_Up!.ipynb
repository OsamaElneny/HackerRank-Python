# Solution:

```python
from collections import deque

def can_pile(cubes):

    d = deque(cubes)
    top = float('inf')  

    while d:
        left = d[0]
        right = d[-1]

        if left <= top and right <= top:
            if left >= right:
                top = d.popleft()
            else:
                top = d.pop()

        elif left <= top:
            top = d.popleft()

        elif right <= top:
            top = d.pop()

        else:
            return False

    return True


t = int(input().strip())
for _ in range(t):
    n = int(input().strip())
    cubes = list(map(int, input().split()))
    print("Yes" if can_pile(cubes) else "No")
```
