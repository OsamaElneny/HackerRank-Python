# Solution:

```python
from collections import namedtuple

def calculate_average_marks(n, columns, data):
    Student = namedtuple("Student", columns)
    students = [Student(*row) for row in data]
    total_marks = sum(int(student.MARKS) for student in students)
    average = total_marks / n
    return round(average, 2)

n = int(input().strip())                    
columns = input().strip().split()            
data = [input().strip().split() for _ in range(n)] 

print(f"Students Average Mark is : {calculate_average_marks(n, columns, data):.2f}")
```
