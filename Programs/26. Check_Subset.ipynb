```python
def check_subset(test_cases):
   
    results = []
    for a, b in test_cases:
        results.append(a.issubset(b))
    return results


t = int(input().strip())
for _ in range(t):
    _ = int(input().strip())  # size of set A (not needed)
    set_a = set(map(int, input().split()))
    _ = int(input().strip())  # size of set B (not needed)
    set_b = set(map(int, input().split()))
    print(set_a.issubset(set_b))
```
