# Solution:

```python
import numpy as np
np.set_printoptions(legacy = "1.13")

N, M = map(int, input().split())
A = np.array([input().split() for _ in range(N)], dtype = int)

print(np.mean(A, axis=1))   # Mean along rows
print(np.var(A, axis=0))    # Variance along columns
print(np.std(A))            # Standard deviation of entire array
```
