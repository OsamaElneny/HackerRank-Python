# Solution:

```python
from collections import OrderedDict

def word_order(n, words):
  
    counts = OrderedDict()

    for word in words:
        counts[word] = counts.get(word, 0) + 1

    distinct_count = len(counts)
    occurrences = list(counts.values())

    return distinct_count, occurrences

n = int(input().strip())
words = [input().strip() for _ in range(n)]

distinct_count, occurrences = word_order(n, words)

print(distinct_count)
print(*occurrences)
```
